```python
import math
import random
import sys
import os

import neat
import pygame
```

```python
# 1️⃣ 创建窗口
screen = pygame.display.set_mode((800, 600))   # 窗口尺寸
pygame.display.set_caption("显示 Sprite 示例")
```

```python
# 2️⃣ 加载图片
sprite = pygame.image.load('car.png').convert()
sprite = pygame.transform.scale(sprite, (120, 120))  # 缩放可选
```

```python
# 3️⃣ 获取矩形位置（方便定位）
rect = sprite.get_rect(center=(400, 300))  # 图片居中
```

```python
clock = pygame.time.Clock()

# 4️⃣ 主循环
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # 5️⃣ 清屏（背景）
    screen.fill((30, 30, 30))  # 深灰背景

    # 6️⃣ 画图（blit = copy image to screen）
    screen.blit(sprite, rect)

    # 7️⃣ 刷新屏幕（显示更新）
    pygame.display.flip()

    clock.tick(60)   # 限制为 60 FPS人·

pygame.quit()
sys.exit()
```

```python
from PIL import Image

def binarize_png(input_path, output_path, threshold=128):
    """
    把一张带灰度的 PNG 变成黑白二值图：
    - 大于阈值的部分全白
    - 否则全黑
    - 保留 alpha 通道
    """
    img = Image.open(input_path).convert("RGBA")
    pixels = img.load()

    for y in range(img.height):
        for x in range(img.width):
            r, g, b, a = pixels[x, y]
            # 计算亮度 (简单平均或加权平均都行)
            gray = (r + g + b) / 3
            if gray >= threshold:
                pixels[x, y] = (255, 255, 255, a)
            else:
                pixels[x, y] = (0, 0, 0, a)

    img.save(output_path)
    print(f"✅ 保存完成: {output_path}")

# 用法示例
binarize_png("K1_Real.png", "K1_Real.png", threshold=180)
```
