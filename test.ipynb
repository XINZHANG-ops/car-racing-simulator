```python
import math
import random
import sys
import os

import neat
import pygame
```

```python
# 1️⃣ 创建窗口
screen = pygame.display.set_mode((800, 600))   # 窗口尺寸
pygame.display.set_caption("显示 Sprite 示例")
```

```python
# 2️⃣ 加载图片
sprite = pygame.image.load('car.png').convert()
sprite = pygame.transform.scale(sprite, (120, 120))  # 缩放可选
```

```python
# 3️⃣ 获取矩形位置（方便定位）
rect = sprite.get_rect(center=(400, 300))  # 图片居中
```

```python
clock = pygame.time.Clock()

# 4️⃣ 主循环
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # 5️⃣ 清屏（背景）
    screen.fill((30, 30, 30))  # 深灰背景

    # 6️⃣ 画图（blit = copy image to screen）
    screen.blit(sprite, rect)

    # 7️⃣ 刷新屏幕（显示更新）
    pygame.display.flip()

    clock.tick(60)   # 限制为 60 FPS人·

pygame.quit()
sys.exit()
```

```python
import pygame

def color_from_index(idx: int, sat=65, val=100):
    """
    用 Golden Angle 生成分散的色相：index 相同 -> 颜色恒定；不同 -> 大概率不同色。
    返回 (r,g,b)
    """
    hue = (idx * 137) % 360  # 137° 接近黄金角，颜色分散好
    c = pygame.Color(0, 0, 0, 255)
    c.hsva = (hue, sat, val, 100)
    return (c.r, c.g, c.b)
```
